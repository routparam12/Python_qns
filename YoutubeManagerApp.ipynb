```python
import json

def load_data():
  try:
    with open("videos.json", "r") as file:    #It will convert the data to json and load them
      test = json.load(file)
      # print(type(test))
      return test
  except FileNotFoundError:
    return []


def save_data(videos):
  with open("videos.json", "w") as file:
    json.dump(videos, file, indent=4)
    # print("Data saved successfully")

# def list_all_videos(videos):
#   for index, videos in enumerate(videos, start = 1):
#     print(f"{index}.")


# def list_all_videos(videos):
#   for vid in videos:
#     print(f"{vid}.")

def list_all_videos(videos):
  print("\n")
  print("*" * 70)
  for index, videos in enumerate(videos, start = 1):
    print(f"{index}.{videos['title']}, Url: {videos['url']}")

def add_videos(videos):
  title = input ("Enter video title")
  url = input ("Enter video url")
  videos.append({"title": title, "url": url})
  save_data(videos)
  print("Video added successfully")

def update_videos(videos):
  list_all_videos(videos)
  index = int(input("Enter video number you waant to update:"))
  if 1 <= index <= len(videos):
    title = input("Enter video name:")
    url = input("Enter video url:")
    videos[index - 1] = {"title": title, "url": url}
    save_data(videos)
    print("Video updated successfully")
  else :
    print("Invalid video number")

def delete_videos(videos):
  list_all_videos(videos)
  index = int(input("Enter video number you want to delete:"))
  if 1 <= index <= len(videos):
    del videos[index - 1]
    save_data(videos)
  else:
    print("invalid vid nuber")
  print("Video deleted successfully")
def main():
  videos = load_data()


  while True:
    print("\n Youtube Manager | Choose an option")
    print("\n 1. List all yt videos")
    print("\n 2. Add a yt video")
    print("\n 3.Update a yt video")
    print("\n 4. Delete a yt video")
    print("\n 5.Exit the app")
    choice = input("Enter your choice: ")
    # print(videos)

    match choice:
      case '1':
        list_all_videos(videos)
      case '2':
        add_videos(videos)
      case '3':
        update_videos(videos)
      case '4':
        delete_videos(videos)
      case '5':
          break
      case _:
        print("Invalid choice")


if __name__ == "__main__":
  main()

```

```python

import sqlite3

connection = sqlite3.connect('ytmanager.db')
# print("Opened database successfully")
cur = connection.cursor()

cur.execute('''
create table if not exists videos(
  Title text not null,
  UrlLink text not null, -- Fixed: changed 'nnot' to 'not' and added a comma
  Length decimal null
)
''')

def list_all_videos(videos):
  cur.execute("SELECT * FROM videos")
  for row in cur.fetchall():
    print(row)

def add_videos(Title, UrlLink, Length):
  cur.execute("INSERT INTO videos (Title, UrlLink, Length) VALUES (?, ?, ?)", (Title, UrlLink, Length))
  connection.commit() # Changed cur.commit() to connection.commit()
  print("Video added successfully")


def update_videos(Id, New_Title, New_UrlLink, New_Length):
   cur.execute("UPDATE videos SET Title = ?, UrlLink = ?, Length = ? WHERE Id = ?", (New_Title, New_UrlLink, New_Length, Id))
   connection.commit() # Changed cur.commit() to connection.commit()
   print("Video updated successfully")


def delete_videos(Id): # Fixed: changed 'videos' parameter to 'Id'
   cur.execute("DELETE FROM videos WHERE Id = ?", (Id,))
   connection.commit() # Changed cur.commit() to connection.commit()
   print("Video deleted successfully")



def main():
  while True:
    print("Youtube Manager with Db| Choose an option")
    print("1. List all yt videos")
    print("2. Add a yt video")
    print("3. Update a yt video")
    print("4. Delete a yt video")
    print("5. Exit the app")
    choice = input("Enter your choice: ")
    # print(videos)

    if choice == '1':
      list_all_videos(None) # Fixed: passing None as videos is not used by the function

    elif choice  == '2':
      Title = input("Add a yt video:")
      UrlLink = input("Link to the video:")
      Length = input("What is the length of the video:")
      add_videos(Title, UrlLink, Length)

    elif choice == '3':
      Id = input("What's the video Id to update ?" )
      Title = input("Update yt video Title")
      UrlLink = input("Update yt video UrlLink")
      Length = input("Update yt video Length")
      update_videos(Id, Title, UrlLink, Length)

    elif choice == '4':
      Id = input("What's the video Id to delete ?")
      delete_videos(Id)

    elif choice == '5':
      break

    else:
      print("Invalid choice")

  connection.close()
  print("Closed database successfully")


if __name__ == "__main__":
  main()
```
