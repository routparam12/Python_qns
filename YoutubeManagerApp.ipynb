```python
import json

def load_data():
  try:
    with open("videos.json", "r") as file:    #It will convert the data to json and load them
      test = json.load(file)
      # print(type(test))
      return test
  except FileNotFoundError:
    return []


def save_data(videos):
  with open("videos.json", "w") as file:
    json.dump(videos, file, indent=4)
    # print("Data saved successfully")

# def list_all_videos(videos):
#   for index, videos in enumerate(videos, start = 1):
#     print(f"{index}.")


# def list_all_videos(videos):
#   for vid in videos:
#     print(f"{vid}.")

def list_all_videos(videos):
  print("\n")
  print("*" * 70)
  for index, videos in enumerate(videos, start = 1):
    print(f"{index}.{videos['title']}, Url: {videos['url']}")

def add_videos(videos):
  title = input ("Enter video title")
  url = input ("Enter video url")
  videos.append({"title": title, "url": url})
  save_data(videos)
  print("Video added successfully")

def update_videos(videos):
  list_all_videos(videos)
  index = int(input("Enter video number you waant to update:"))
  if 1 <= index <= len(videos):
    title = input("Enter video name:")
    url = input("Enter video url:")
    videos[index - 1] = {"title": title, "url": url}
    save_data(videos)
    print("Video updated successfully")
  else :
    print("Invalid video number")

def delete_videos(videos):
  list_all_videos(videos)
  index = int(input("Enter video number you want to delete:"))
  if 1 <= index <= len(videos):
    del videos[index - 1]
    save_data(videos)
  else:
    print("invalid vid nuber")
  print("Video deleted successfully")
def main():
  videos = load_data()


  while True:
    print("\n Youtube Manager | Choose an option")
    print("\n 1. List all yt videos")
    print("\n 2. Add a yt video")
    print("\n 3.Update a yt video")
    print("\n 4. Delete a yt video")
    print("\n 5.Exit the app")
    choice = input("Enter your choice: ")
    # print(videos)

    match choice:
      case '1':
        list_all_videos(videos)
      case '2':
        add_videos(videos)
      case '3':
        update_videos(videos)
      case '4':
        delete_videos(videos)
      case '5':
          break
      case _:
        print("Invalid choice")


if __name__ == "__main__":
  main()

```
