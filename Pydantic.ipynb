```python
from pydantic import BaseModel,EmailStr, field_validator
from typing import List
```

```python
# First, install the necessary email-validator package for Pydantic.
# This is required for the EmailStr type validation.


class PydanticPatient(BaseModel):
    name: str
    email: EmailStr
    allergies: List[str]
    contact : dict[str,str]
    age:int

    @field_validator('email')
    @classmethod
    def email_validator(cls, value):
      valid_domain = ['hdfc.com','icici.com']
      domain_name = value.split('@')[-1]
      if domain_name not in valid_domain:
          raise ValueError('Invalid email domain')
      return value

    @field_validator('name')
    @classmethod
    def transform_name(cls,value):
      return value.upper()

    @field_validator('age', mode = 'after')
    @classmethod
    def check_age(cls, value):
      if 0 < value < 100:
        return value
      else:
        raise ValueError('Age cannot be negative')

      return value


def update_patient(patient: PydanticPatient):
  print(patient.name)
  print(patient.email)

patient_info = {'name':'Param', 'email':'Param@hdfc.com', 'allergies':['pollen','dust'], 'contact':{'Email':'parm@gmail.com','Number':'9999888'}, 'age': 12}

patient1 = PydanticPatient(**patient_info)
update_patient(patient1)
```

#Model validator
It can check multiple fields at a time like, if the age is > 60,there should be an emegency number in the contact details . if both satisfied then only user is allowed to make his profile .

```python
from pydantic import BaseModel,EmailStr, model_validator
from typing import List
```

```python
class PydanticPatient(BaseModel):
    name: str
    email: EmailStr
    allergies: List[str]
    contact : dict[str,str]
    age:int

    @model_validator(mode='after')
    def validate_emergency_number(self):
      if self.age > 60 and 'emergency' not in self.contact:
        raise ValueError('Emergency number is required')
      return self


def update_patient(patient: PydanticPatient):
  print(patient.name)
  print(patient.email)

patient_info = {'name':'Param', 'email':'Param@hdfc.com', 'allergies':['pollen','dust'], 'contact':{'Email':'parm@gmail.com','Number':'9999888'}, 'age': 55}
# hage should be less than 60, otherwise a emergency number is required
patient1 = PydanticPatient(**patient_info)
update_patient(patient1)
```

#computed field


```python
from pydantic import BaseModel,EmailStr, model_validator, computed_field
from typing import List

class patientDetails(BaseModel):

  name: str
  email:EmailStr
  age:int
  weight:float
  height:float
  married:bool
  allergies:List[str]
  contactdetails: dict[str, str]

  @computed_field
  @property
  def calculate_bmi(self) -> float:
    bmi = round(self.weight / (self.height ** 2), 2)
    return bmi

def update_paient_information(patient: patientDetails):
  print(patient.name)
  print(patient.email)
  print(patient.calculate_bmi)


patient_info = {'name':'Param','email':'param@gmail.com', 'age':'23','height':1.78,'weight':'80','married':'True','allergies':['pollen','dust'],'contactdetails':{'Email':'parm@gmail.com','Number':'9999888'}}

patient1 = patientDetails(**patient_info)
update_paient_information(patient1)
```

#Nested Model

```python
from pydantic import BaseModel,EmailStr, model_validator, computed_field
from typing import List
```

```python
class Address(BaseModel):
  city:str
  state:str
  country:str
  pin:int


class Patients(BaseModel):
  name:str
  email:EmailStr
  age:int
  address: Address

address_dict = {'city':'Gurgaon','state':'Haryana','country':'India','pin':'122001'}
address1 = Address(**address_dict)
patient_data = {'name':'Param','email':'param@gmail.com','age':23,'address':address1}
patient1 = Patients(**patient_data)
print(patient1)
print(patient1.name)
print(patient1.address.city)

temp = patient1.model_dump()    #Convert the current file into dictonary
print(temp)
print(type(temp))

temp1 = patient1.model_dump(exclude={'address':['state']})    #Convert the current file into dictonary and exvludes declared variable
print(temp1)
print(type(temp1))

temp2 = patient1.model_dump_json()  #Convert the current file into json
print(temp2)
print(type(temp2))
```
